```python
import pandas as pd #import pandas
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0] 
df
```

Imported the data from wikipedia link into a pandas dataframe above

Below the cells with a borough 'Not assigned' are ignored

```python
# ignore cells with a borough that is Not assigned

import numpy as np

df.replace("Not assigned", np.nan, inplace = True)
df.dropna(axis=0, inplace=True)
df
```

This is the final dataframe and .shape method is used below to print the number of rows and columns

```python
df.shape
```

The csv with geospatial data is imported below and the dataframes are merged (as geocoder would not import)

```python
df_longlat = pd.read_csv('http://cocl.us/Geospatial_data')
df_longlat
```

```python
df_combined = pd.merge(df, df_longlat, on='Postal Code')
df_combined
```

```python
toronto_data = df_combined[df_combined['Borough'].astype(str).str.contains('Toronto')]
toronto_data.head()
```

First 5 rows of the dataframe relating to Toronto

```python
from geopy.geocoders import Nominatim

address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinates of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
!pip install folium
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors

# create map
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Borough'], toronto_data['Neighborhood']):
    label = '{}, {}'.format('Neighborhood', 'Borough')
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [latitude, longitude],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.5,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
